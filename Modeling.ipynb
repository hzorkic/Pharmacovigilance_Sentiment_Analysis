```python
#-------------------------------------------------#
# look at top 10 most positive and negative words #
#-------------------------------------------------#

def get_counts(x_test, y_pred, y_test):
  # convert to dataframe to add prediction column
  testdf = pd.DataFrame(x_test)
  # convert predictions to list
  y_pred_list = y_pred.tolist()
  # convert actual y's to list
  y_test_list = y_test.tolist()
  # add predicted y's to df
  testdf = testdf.assign(pred_polarity=y_pred_list)
  # add actual y's to df
  testdf = testdf.assign(test_polarity=y_test_list)
  
  #now we have a dataframe with columns review, pred_polarity, test_polarity

  token_reviews = []
  for r in testdf.review:
    token_r = []
    stokenized = sent_tokenize(r)
    for sent in stokenized:
      words = word_tokenize(sent)
      token_r.extend(words)
    token_reviews.append(token_r)

  testdf['tokenized_reviews']=(token_reviews)

  #filter polarity to lists
  pos_words = testdf['tokenized_reviews'][(testdf['pred_polarity'] == 1)].to_list()
  neg_words = testdf['tokenized_reviews'][(testdf['pred_polarity'] == 0)].to_list()

  # adding words to dictionary if it's just letters--no numbers or symbols
  pos = dict()
  for i in pos_words:
    for word in i:
      if word.isalpha() == True:
        pos[word] = pos.get(word,0)+1
  neg = dict()
  for i in neg_words:
    for word in i:
      if word.isalpha() == True:
        neg[word] = neg.get(word,0)+1
  
  # nest positive and negative dictionaries
  counts={'pos':pos,'neg':neg}
  
  return counts
  
#counts = get_counts(x_test, y_pred, y_test)

def get_word_prob(counts, word, polarity):
  if word not in counts[polarity]:
    return 0.0
    
  # find total number of events
  pos_events = sum(counts["pos"].values())
  neg_events = sum(counts["neg"].values())
  total_events = pos_events + neg_events
    
  # find number of events of word in positive and negative polarities
  if word in counts["pos"]:
    word_pos_events = counts["pos"][word]
  else:
    word_pos_events = 0
  if word in counts["neg"]:
    word_neg_events = counts["neg"][word]
  else:
    word_neg_events = 0
  word_total_events = word_pos_events + word_neg_events
    
  #probability of polarity given the word
  polarity_g_word = counts[polarity][word]/word_total_events
  #probability of the word out of all the words
  prob_word = word_total_events / total_events
  #probability of the polarity out of all events
  prob_polarity = sum(counts[polarity].values()) / total_events
    
  probability = (polarity_g_word * prob_word / prob_polarity)
    
    
  return probability # P(word|polarity)

def log_odds_ratio(counts, word, polarity):
  if polarity == "pos":
    inverse_polarity = "neg"
  else:
    inverse_polarity = "pos"

  # probability of word given the given polarity and opposite polarity
  prob_word_g_pol = get_word_prob(counts, word, polarity)
  prob_word_g_ipol = get_word_prob(counts, word, inverse_polarity)
  if prob_word_g_ipol == 0:
    return 10000
  if prob_word_g_pol == 0:
    return 0
  log_odds_ratio = math.log((prob_word_g_pol/prob_word_g_ipol))
  return log_odds_ratio

def sort_pos_words(x_test, y_pred, y_test):
    import copy
    counts = get_counts(x_test, y_pred, y_test)
    counts10 = {"pos":{}, "neg":{}}

    def ten_times(word): #changed to 50 instead of 10
      # find how many times the word shows up in the POSITIVE sample space
      if word in counts_original["pos"]:
        word_pos_sample_space = counts_original["pos"][word]
      else:
        word_pos_sample_space = 0
      # find how many times the word shows up in the NEGATIVE sample space
      if word in counts_original["neg"]:
        word_neg_sample_space = counts_original["neg"][word]
      else:
        word_neg_sample_space = 0
            
      # add pos + neg occurances
      word_sample_space = word_neg_sample_space + word_pos_sample_space
       
      if (word_sample_space > 50 ):
      # passes the test
        return True, word_sample_space
      else:
      # does not pass the test
        return False, word_sample_space

    # make a copy of the counts data
    counts_original = get_counts(x_test, y_pred, y_test)
    
    # {"pos":{word:count, word:count, ...}, "neg":{word:count, word:count, ...}}
    counts10 = copy.deepcopy(counts_original)
    
    # for every word in the positive dataset
    for key, val in counts_original["pos"].items():
      # if if exists 
      if key in counts10["pos"]:
        ten_check, counts = ten_times(key)
        # check if this word appears at least 10 times in both positive and negative datasets
        if (ten_check == False):
          # del the word from the positive sample space IF it fails the ten test 
          del counts10["pos"][key] 
          # if this word also negative space, delete it
          if key in counts10["neg"]:
           del counts10["neg"][key] 
            
    for key, val in counts_original["neg"].items():
      if key in counts10["neg"]:
        ten_check, counts = ten_times(key)
        if (ten_check == False):
          del counts10["neg"][key] 
          if key in counts10["pos"]:
            del counts10["pos"][key] 
    
    unsorted_list = []
    
    for word in counts10["pos"]:
      word_lor = [word, log_odds_ratio(counts10, word, "pos")]
      unsorted_list.append(tuple(word_lor))
    
    sorted_list = sorted(unsorted_list, key=lambda x: x[1], reverse=True) 

    return sorted_list 
   
```
